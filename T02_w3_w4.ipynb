5. –î–∞–Ω—ã –¥–≤–∞ –æ—Ç—Ä–µ–∑–∫–∞  [ùëé;ùëè]  –∏  [ùëê;ùëë] . –ù–∞–π–¥–∏—Ç–µ –∏—Ö –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ. –ï—Å–ª–∏ –æ—Ç—Ä–µ–∑–∫–∏ –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è, —Ç–æ –≤—ã–¥–∞–π—Ç–µ —Å–æ–æ–±—â–µ–Ω–∏–µ.

````python
def find_intersection(A, B, C, D):
    # –í—ã—á–∏—Å–ª—è–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏—è –ø—Ä—è–º—ã—Ö –ø—Ä–æ—Ö–æ–¥—è—â–∏—Ö —á–µ—Ä–µ–∑ AB –∏ CD
    def line_equation(point1, point2):
        x1, y1 = point1
        x2, y2 = point2
        a = y2 - y1
        b = x1 - x2
        c = a * x1 + b * y1
        return a, b, c

    a1, b1, c1 = line_equation(A, B)
    a2, b2, c2 = line_equation(C, D)

    # –ù–∞—Ö–æ–¥–∏–º —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è
    determinant = a1 * b2 - a2 * b1

    if determinant == 0:
        # –û—Ç—Ä–µ–∑–∫–∏ –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã, –Ω–µ—Ç —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è
        return None
    else:
        x = (c1 * b2 - c2 * b1) / determinant
        y = (a1 * c2 - a2 * c1) / determinant
        return x, y

# –ó–∞–¥–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–µ–∫ A, B, C –∏ D

A1 = int(input())
A2 = int(input())
B1 = int(input())
B2 = int(input())
C1 = int(input())
C2 = int(input())
D1 = int(input())
D2 = int(input())

A = [A1, A2]
B = [B1, B2]
C = [C1, C2]
D = [D1, D2]

print(f"–û—Ç—Ä–µ–∑–æ–∫ –ê–í: –ê[{A1}, {A2}], B[{B1}, {B2}]\n–û—Ç—Ä–µ–∑–æ–∫ CD: C[{C1}, {C2}], D[{D1}, {D2}]")

# –ù–∞—Ö–æ–¥–∏–º —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è
intersection = find_intersection(A, B, C, D)

if intersection:
    print("–¢–æ—á–∫–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è:", intersection)
else:
    print("–û—Ç—Ä–µ–∑–∫–∏ AB –∏ CD –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã, –Ω–µ—Ç —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è.")
````

6. –ü–æ –Ω–æ–º–µ—Ä—É –≥–æ–¥–∞ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –¥–∞–Ω–Ω—ã–π –≥–æ–¥ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º. (–≥–æ–¥ —è–≤–ª—è–µ—Ç—Å—è –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º, –µ—Å–ª–∏ –µ–≥–æ –Ω–æ–º–µ—Ä –∫—Ä–∞—Ç–µ–Ω 4, –Ω–æ –Ω–µ –∫—Ä–∞—Ç–µ–Ω 100, –∞ —Ç–∞–∫–∂–µ –µ—Å–ª–∏ –æ–Ω –∫—Ä–∞—Ç–µ–Ω 400).

7. –î–ª—è –æ—Ç—Ä–µ–∑–∫–æ–≤ –¥–ª–∏–Ω—ã a, b, c –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –º–æ–∂–Ω–æ –ª–∏ –∏–∑ –Ω–∏—Ö —Å–æ—Å—Ç–∞–≤–∏—Ç—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –∏ —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —ç—Ç–æ—Ç —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω—ã–º.

````python
a = int(input())
b = int(input())
c = int(input())

if a+b>c or a+c>b or b+c>a:
    print("–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç")
else:
    print("–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç")

if a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
    print ("–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω—ã–π")
else:
    print("–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏ –Ω–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω—ã–π")
````

8. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è —Ä–µ—à–µ–Ω–∏—è —É—Ä–∞–≤–Ω–µ–Ω–∏—è
$$ ax^2+bx+c=0 $$
–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã $a, b, c$ –º–æ–≥—É—Ç –±—ã—Ç—å –ª—é–±—ã–º–∏ —á–∏—Å–ª–∞–º–∏. 

````python
a = int(input())
b = int(input())
c = int(input())

D = (b**2)-(4*a*c)

if D>0:
    x1 = (-b(D**(1/2)))/(2*a)
    x2 = (-b(D**(1/2)))/(2*a)
    print(f"–î–≤–∞ –∫–æ—Ä–Ω—è: {x1} –∏ {x2}")
elif D==0:
    x = (-b)/(2*a)
    print("–û–¥–∏–Ω –∫–æ—Ä–µ–Ω—å:", x)
else:
    print("–ù–µ—Ç –∫–æ—Ä–Ω–µ–π")
````

12. –í—ã—á–∏—Å–ª–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä–æ–µ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –∏ –∑–Ω–∞–∫–æ–≤ ¬´+¬ª –∏ ¬´-¬ª. –í—ã—Ä–∞–∂–µ–Ω–∏–µ –≤–≤–æ–¥–∏—Ç—Å—è –∫–∞–∫ —Å–∏–º–≤–æ–ª—å–Ω–∞—è —Å—Ç—Ä–æ–∫–∞.

````python
expression = input("–í–≤–µ–¥–∏—Ç–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ: ") # –ø—Ä–∏–Ω–∏–º–∞–µ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è

# —É–¥–∞–ª—è–µ–º –ø—Ä–æ–±–µ–ª—ã –∏–∑ –≤—ã—Ä–∞–∂–µ–Ω–∏—è
expression = expression.replace(" ", "")

# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
result = 0
operator = '+'
current_number = ''

# –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –∫–∞–∂–¥—ã–π —Å–∏–º–≤–æ–ª –≤ –≤—ã—Ä–∞–∂–µ–Ω–∏–∏
for char in expression:
    if char.isdigit():
        current_number += char
    else:
        # –µ—Å–ª–∏ —á–∏—Å–ª–æ –±—ã–ª–æ –Ω–∞–∫–æ–ø–ª–µ–Ω–æ, —Ç–æ –¥–æ–±–∞–≤–ª—è–µ–º –µ–≥–æ –∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
        if current_number:
            if operator == '+':
                result += int(current_number)
            elif operator == '-':
                result -= int(current_number)
            
            current_number = '' # —Å–±—Ä–∞—Å—ã–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —á–∏—Å–ª–∞
            
        # –æ–±–Ω–æ–≤–ª—è–µ–º –∑–Ω–∞–∫ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞
        operator = char

# –¥–æ–±–∞–≤–ª—è–µ–º –ø–æ—Å–ª–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É (–µ—Å–ª–∏ –µ—Å—Ç—å)
if current_number:
    if operator == '+':
        result += int(current_number)
    elif operator == '-':
        result -= int(current_number)

print("–†–µ–∑—É–ª—å—Ç–∞—Ç: ", result)
````

14. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª —Å—Ç—Ä–æ–∫—É —Å–ª–µ–¥—É—é—â–µ–≥–æ –≤–∏–¥–∞: ¬´a op b¬ª, –≥–¥–µ a –∏ b ‚Äì –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —á–∏—Å–ª–∞, –∞ ‚Äòop‚Äô –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä –∏ –º–æ–∂–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è  ¬´+,‚àí,‚àó,/,‚àó‚àó, . –í –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ —Å –ø–æ–º–æ—â—å—é —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —Å—Ç—Ä–æ–∫ –≤—ã–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ: ¬´a + b = 3 + 2 = 5¬ª, –≥–¥–µ –±—ã–ª–∞ –ø–æ–ª—É—á–µ–Ω–∞ —Å—Ç—Ä–æ–∫–∞ ¬´3 + 2¬ª. –î–ª—è –æ–ø–µ—Ä–∞—Ü–∏–π –≤–æ–∑–≤–µ–¥–µ–Ω–∏—è –≤ —Å—Ç–µ–ø–µ–Ω—å –∏ –¥–µ–ª–µ–Ω–∏—è –ø–æ –º–æ–¥—É–ª—é –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤–º–µ—Å—Ç–æ –∑–Ω–∞–∫–∞ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è. –ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —Å—Ç—Ä–æ–∫–∏ ¬´2 ** 3¬ª –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤—ã–≤–µ–¥–µ–Ω–æ ¬´a –≤ —Å—Ç–µ–ø–µ–Ω–∏ b = 2 –≤ —Å—Ç–µ–ø–µ–Ω–∏ 3 = 8¬ª.

````python
# –≤–≤–æ–¥ —Å—Ç—Ä–æ–∫–∏
user_input = input("–í–≤–µ–¥–∏—Ç–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 'a op b'): ")

# —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –≤–≤–æ–¥ –Ω–∞ —á–∏—Å–ª–∞ –∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä
parts = user_input.split() #—Ä–∞–∑–¥–µ–ª—è–µ–º –ø–æ –ø—Ä–æ–±–µ–ª–∞–º –≤ —Å–ø–∏—Å–æ–∫
a = float(parts[0])
op = parts[1]
b = float(parts[2])

# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞
if op == '+':
    result = a + b
    operator_string = "+"
elif op == '-':
    result = a - b
    operator_string = "-"
elif op == '*':
    result = a * b
    operator_string = "*"
elif op == '/':
    result = a / b
    operator_string = "/"
elif op == '**':
    result = a ** b
    operator_string = "–≤ —Å—Ç–µ–ø–µ–Ω–∏"
else:
    result = "–ù–µ–≤–µ—Ä–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä"

# –≤—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ —Å —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ–º
if result != "–ù–µ–≤–µ—Ä–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä":
    print(f"{a} {operator_string} {b} = {result}")
else:
    print(result)
````

16. –ò–º–µ—é—Ç—Å—è –¥–≤–∞ n ‚Äì –º–µ—Ä–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞ x –∏ y, –∫–æ—Ç–æ—Ä—ã–µ –∑–∞–¥–∞—é—Ç –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã n —Ç–æ—á–µ–∫ –Ω–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ (—Å–ª—É—á–∞–π–Ω—ã–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞). –ù–∞–π—Ç–∏ –Ω–∞–∏–±–æ–ª–µ–µ –±–ª–∏–∑–∫–∏–µ –¥—Ä—É–≥ –¥—Ä—É–≥—É —Ç–æ—á–∫–∏.

````python
import numpy as np

# —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å n
n = int(input("–í–≤–µ–¥–∏—Ç–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å n: "))

# –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã n —Ç–æ—á–µ–∫ –¥–ª—è –≤–µ–∫—Ç–æ—Ä–∞ x
x = np.array([int(input(f"–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É {i+1} –¥–ª—è –≤–µ–∫—Ç–æ—Ä–∞ x: ")) for i in range(n)])

# –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã n —Ç–æ—á–µ–∫ –¥–ª—è –≤–µ–∫—Ç–æ—Ä–∞ y
y = np.array([int(input(f"–í–≤–µ–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—É {i+1} –¥–ª—è –≤–µ–∫—Ç–æ—Ä–∞ y: ")) for i in range(n)])

# —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏
distance = np.linalg.norm(x - y)

print("–í–µ–∫—Ç–æ—Ä x:", x)
print("–í–µ–∫—Ç–æ—Ä y:", y)
print("–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏:", distance)

# –±–ª–∏–∂–∞–π—à–∏–µ —Ç–æ—á–∫–∏
min_distance_index = np.argmin(np.abs(x - y))
closest_x = x[min_distance_index]
closest_y = y[min_distance_index]

print("–ë–ª–∏–∂–∞–π—à–∏–µ —Ç–æ—á–∫–∏:", closest_x, closest_y)
````

17. –ò–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ N –≤—ã–¥–µ–ª–∏—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤–æ N2 —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ã–µ 2, –º–Ω–æ–∂–µ—Å—Ç–≤–æ N3 ‚Äì –∫—Ä–∞—Ç–Ω—ã–µ 3, –º–Ω–æ–∂–µ—Å—Ç–≤–æ N 6 ‚Äì –∫—Ä–∞—Ç–Ω—ã–µ 6 (—Ç.–µ. –∫—Ä–∞—Ç–Ω—ã–µ –∏ 2 –∏ 3), –º–Ω–æ–∂–µ—Å—Ç–≤–æ N 23 ‚Äì –∫—Ä–∞—Ç–Ω—ã–µ –ª–∏–±–æ 2, –ª–∏–±–æ 3.

````python
N = int(input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ N: "))

# –º–Ω–æ–∂–µ—Å—Ç–≤–æ N2 - —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ã–µ 2
N2 = set()
for i in range(1, N + 1):
    if i % 2 == 0:
        N2.add(i)

# –º–Ω–æ–∂–µ—Å—Ç–≤–æ N3 - —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ã–µ 3
N3 = set()
for i in range(1, N + 1):
    if i % 3 == 0:
        N3.add(i)

# –º–Ω–æ–∂–µ—Å—Ç–≤–æ N6 - —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ã–µ –∏ 2 –∏ 3
N6 = N2.intersection(N3)

# –º–Ω–æ–∂–µ—Å—Ç–≤–æ N23 - —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ã–µ –ª–∏–±–æ 2, –ª–∏–±–æ 3
N23 = N2.union(N3)

print("–ú–Ω–æ–∂–µ—Å—Ç–≤–æ N2:", N2)
print("–ú–Ω–æ–∂–µ—Å—Ç–≤–æ N3:", N3)
print("–ú–Ω–æ–∂–µ—Å—Ç–≤–æ N6:", N6)
print("–ú–Ω–æ–∂–µ—Å—Ç–≤–æ N23:", N23)
````

18. –î–∞–Ω—ã m (m>1) —Å–ª–æ–≤. –ù–∞–π—Ç–∏ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–¥–∞–Ω–Ω–æ–π –±—É–∫–≤—ã –≤ —ç—Ç–∏—Ö —Å–ª–æ–≤–∞—Ö.

````python
m = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤ (m > 1): "))
if m <= 1:
    print("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ 1.")
else:
    letter = input("–í–≤–µ–¥–∏—Ç–µ –±—É–∫–≤—É, –∫–æ—Ç–æ—Ä—É—é –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏: ")
    words = []

    for i in range(m):
        word = input(f"–í–≤–µ–¥–∏—Ç–µ —Å–ª–æ–≤–æ {i + 1}: ")
        words.append(word)

    total_count = 0

    for word in words:
        count = word.count(letter)
        total_count += count

    print(f"–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±—É–∫–≤—ã '{letter}' –≤ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö —Å–ª–æ–≤–∞—Ö: {total_count}")
````

19. –î–∞–Ω—ã m (m>1) —Å–ª–æ–≤. –í –∫–∞–∫–æ–º –∏–∑ –Ω–∏—Ö –¥–æ–ª—è (–≤ %) –∑–∞–¥–∞–Ω–Ω–æ–π –±—É–∫–≤—ã –±–æ–ª—å—à–µ.

````python
def calculate_letter_percentage(word, letter): #—Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å–ª–æ–≤–æ –∏ –±—É–∫–≤—É –∏ –≤—ã—á–∏—Å–ª—è–µ—Ç
    count = word.count(letter)                 #–ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –±—É–∫–≤—ã –≤ —Å–ª–æ–≤–µ
    total_characters = len(word)
    percentage = (count / total_characters) * 100
    return percentage

#–ø—Ä–æ—Ö–æ–¥–∏—Ç –ø–æ –∫–∞–∂–¥–æ–º—É —Å–ª–æ–≤—É –≤ —Å–ø–∏—Å–∫–µ –∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç 
#–ø—Ä–µ–¥—ã–¥—É—â—É—é —Ñ—É–Ω–∫—Ü–∏—é, —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ —Å–ª–æ–≤–æ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –¥–æ–ª–µ–π —É–∫–∞–∑–∞–Ω–Ω–æ–π –±—É–∫–≤—ã
def find_word_with_highest_percentage(word_list, letter): #—Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤ –∏ –±—É–∫–≤—É
    max_percentage = 0
    word_with_max_percentage = None

    for word in word_list:
        percentage = calculate_letter_percentage(word, letter)
        if percentage > max_percentage: #–µ—Å–ª–∏ –µ—Å—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –±—É–∫–≤–∞
            max_percentage = percentage
            word_with_max_percentage = word
        print(max_percentage,"%", word_with_max_percentage)
    return word_with_max_percentage

m = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤ (m > 1): "))
word_list = []

for i in range(m):
    word = input(f"–í–≤–µ–¥–∏—Ç–µ —Å–ª–æ–≤–æ {i + 1}: ")
    word_list.append(word)

target_letter = input("–í–≤–µ–¥–∏—Ç–µ –±—É–∫–≤—É, –¥–æ–ª—é –∫–æ—Ç–æ—Ä–æ–π –≤—ã —Ö–æ—Ç–∏—Ç–µ –Ω–∞–π—Ç–∏: ")

result_word = find_word_with_highest_percentage(word_list, target_letter)

if result_word:
    print(f"–°–ª–æ–≤–æ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –¥–æ–ª–µ–π –±—É–∫–≤—ã '{target_letter}' - {result_word}")
else:
    print(f"–ù–∏ –≤ –æ–¥–Ω–æ–º –∏–∑ —Å–ª–æ–≤ –¥–æ–ª—è –±—É–∫–≤—ã '{target_letter}' –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 0%.")
````
