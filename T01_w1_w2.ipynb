````python
x1, y1 = 1, 1
x2, y2 = 5, 1
x3, y3 = 3, 3

a = ( ((x2-x1)**2)+((y2-y1)**2) )**(1/2)
b = ( ((x3-x2)**2)+((y3-y2)**2) )**(1/2)
c = ( ((x3-x1)**2)+((y3-y1)**2) )**(1/2)

p = a + b + c
pp = p/2
s = (pp*(pp-a)*(pp-b)*(pp-c))**(1/2)

print (a, b, c)
print("Периметр треугольника равен ", round(p,3), ", площадь треугольника равна ", round(s,3), sep = "")
````

1.12 Пусть пользователь вводит 5 чисел: а, b, c, d, e. Реализуйте программу расчета выражения вида $res=\frac{a+bc}{2d-e}+Ost[b^e/c]$, где $Ost$ - остаток от деления. Учесть невозможность деления на 0.

````python
import sys
a = int(input("Введите значение a: "))
b = int(input("Введите значение b: "))
c = int(input("Введите значение c: "))
if c == 0:
    sys.exit("Некорректное значение переменной c. Деление на ноль.")
d = int(input("Введите значение d: "))
e = int(input("Введите значение e: "))
part1 = a+b*c
part2 = 2*d-e
if part2 == 0:
    sys.exit("Некорректные значения. Деление на ноль при вычислении.")
res = (part1/part2) + ((b**e)%c)
print("Результат:", res)
````

````python
def find_intersection(A, B, C, D):
    # Вычисляем уравнения прямых проходящих через AB и CD
    def line_equation(point1, point2):
        x1, y1 = point1
        x2, y2 = point2
        a = y2 - y1
        b = x1 - x2
        c = a * x1 + b * y1
        return a, b, c

    a1, b1, c1 = line_equation(A, B)
    a2, b2, c2 = line_equation(C, D)

    # Находим точку пересечения
    determinant = a1 * b2 - a2 * b1

    if determinant == 0:
        # Отрезки параллельны, нет точки пересечения
        return None
    else:
        x = (c1 * b2 - c2 * b1) / determinant
        y = (a1 * c2 - a2 * c1) / determinant
        return x, y

# Задаем координаты точек A, B, C и D
A = (1, 1)
B = (5, 5)
C = (1, 4)
D = (7, 2)

# Находим точку пересечения
intersection = find_intersection(A, B, C, D)

if intersection:
    print("Точка пересечения:", intersection)
else:
    print("Отрезки AB и CD параллельны, нет точки пересечения.")
````
