4. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–¥–∏—É—Å–∞ r –≤—ã—á–∏—Å–ª—è–µ—Ç –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ –∏ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏. –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ—Ä—Ç–µ–∂ –∏–∑ 2 –∑–Ω–∞—á–µ–Ω–∏–π.

5. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —á–µ—Ç—ã—Ä–µ—Ö –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π, –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—â–∏—Ö —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –≤ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ. –°–æ–∑–¥–∞—Ç—å —Å–ª–æ–≤–∞—Ä—å —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –∏—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –∏–º —Å–∏–º–≤ –æ–ª–∞–º–∏ –æ–ø–µ—Ä–∞—Ü–∏–π. –î–ª—è –¥–≤—É—Ö –∑–∞—Ä–∞–Ω–µ–µ –∑–∞–¥–∞–Ω–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª (–Ω–∞–ø—Ä–∏–º–µ—Ä, 25 –∏ 4) –≤—ã–ø–æ–ª–Ω–∏—Ç—å –≤—ã–±—Ä–∞–Ω–Ω—É—é –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫—É—é –æ–ø–µ—Ä–∞—Ü–∏—é.

7. –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ (–æ—Ç 0 –¥–æ 999) –∏–∑ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ –≤ —á–∏—Å–ª–æ —Ç–∏–ø–∞ int. –ü—Ä–∏–º–µ—Ä: `to_int('—Ç—Ä–∏–¥—Ü–∞—Ç—å —Ç—Ä–∏') -> 33`

8. –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ (–æ—Ç 0 –¥–æ 999) –∏–∑ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ –≤ —á–∏—Å–ª–æ —Ç–∏–ø–∞ int. –ò —Å–æ–æ–±—â–∞–µ—Ç –æ–± –æ—à–∏–±–∫–∞—Ö (–≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –æ–ø–∏—Å–∞–Ω–∏–µ —Ç–∏–ø–∞ –æ—à–∏–±–∫–∏ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∏—Å–ª–æ -1). –ü—Ä–∏–º–µ—Ä: to_int('—Ç—Ä–∏–¥—Ü–∞—Ç—å —Ç—Ä–∏') -> 33 –ü—Ä–∏–º–µ—Ä: to_int("—Å—Ç–æ —Å–æ—Ä–æ–∫ —Ç—Ä–∏–¥—Ü–∞—Ç—å –¥–≤–∞ —Å—Ç–æ") -> -1 –í—ã–≤–æ–¥: —Ç—Ä–∏–¥—Ü–∞—Ç—å - –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ –≤ —á–∏—Å–ª–µ.

1. –ò–º–µ–µ—Ç—Å—è —Å–ø–∏—Å–æ–∫ –∏–∑ 20 —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –æ—Ç 0 –¥–æ 100. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞–∑–±–∏–≤–∞–µ—Ç —ç—Ç–æ—Ç —Å–ø–∏—Å–æ–∫ –Ω–∞  ùëõ  —Å–ø–∏—Å–∫–æ–≤ –ø–æ  ùëò  —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏ –≤–∫–ª—é—á–∞–µ—Ç –≤—Å–µ —ç—Ç–∏ —Å–ø–∏—Å–∫–∏ –≤ –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ  ùëò  –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ —Ñ—É–Ω–∫—Ü–∏–∏.

–í —Ñ—É–Ω–∫—Ü–∏—é split_list –ø–µ—Ä–µ–¥–∞—é—Ç—Å—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: —Å–ø–∏—Å–æ–∫ data_list –∏ –∑–Ω–∞—á–µ–Ω–∏–µ k. k - –¥–ª–∏–Ω–∞ –ø–æ–¥—Å–ø–∏—Å–∫–æ–≤, –Ω–∞ –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥–µ—Ç —Ä–∞–∑–±–∏—Ç –æ—Å–Ω–æ–≤–Ω–æ–π —Å–ø–∏—Å–æ–∫.
–í —Ñ—É–Ω–∫—Ü–∏–∏ —Å–æ–∑–¥–∞–µ—Ç—Å—è —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π —Å–ø–∏—Å–æ–∫ (–∏–∑–Ω–∞—á–∞–ª—å–Ω–æ –ø—É—Å—Ç–æ–π), –ø–æ—Ç–æ–º –≤ —Ü–∏–∫–ª–∫–µ —Å –ø–æ–º–æ—â—å—é –±—É—Ñ–µ—Ä–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π —Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ –∏–∑ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞ –∏ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫ –ø–æ–¥—Å–ø–∏—Å–æ–∫ –≤ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π —Å–ø–∏–∫–æ–≤.

````python
def split_list(data_list, k):
    result = []
    for i in range(0, len(data_list), k):
        sublist = data_list[i:i+k]
        result.append(sublist)
    return result

data = input("–í–≤–µ–¥–∏—Ç–µ 20 —á–∏—Å–µ–ª —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª: ").split()
data = [int(num) for num in data]
n = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–∏—Å–∫–æ–≤ –¥–ª—è —Ä–∞–∑–±–∏–µ–Ω–∏—è: "))

k = len(data) // n
result = split_list(data, k)
print(result)
````

2. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –≤—ã—á–∏—Å–ª—è—é—â—É—é —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞, –ø–µ—Ä–µ–¥–∞–≤–∞–µ–º–æ–≥–æ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞.

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤–æ–¥–∏—Ç —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –ø–µ—Ä–µ–¥–∞–µ—Ç—Å—è –≤ —Ñ—É–Ω–∫—Ü–∏—é Factorial. –í —Ñ—É–Ω–∫—Ü–∏–∏ –≤–≤–æ–¥–∏—Ç—Å—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è-—Å—á–µ—Ç—á–∏–∫ i –∏ –Ω–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ x. –î–∞–ª–µ–µ –≤ —Ü–∏–∫–ª–µ –Ω–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ—à–∞–≥–æ–≤–æ —É–º–Ω–æ–∂–∞–µ—Ç—Å—è –Ω–∞ —Å—á–µ—Ç—á–∏–∫, –∫–æ—Ç–æ—Ä—ã–π —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –¥–æ —Ç–æ–≥–æ –º–æ–º–µ–Ω—Ç–∞, –ø–æ–∫–∞ –Ω–µ –¥–æ—Å—Ç–∏–≥–Ω–µ—Ç –≤–µ–ª–∏—á–∏–Ω—ã —á–∏—Å–ª–∞, –≤–≤–µ–¥–µ–Ω–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º.

````python
def Factorial(num_fact):
    i = 1
    x = 1
    while i<=num_fact:
        x *= i
        i+= 1
    return x

num_factorial = int(input())

result = Factorial(num_factorial)

print(result)
````

3. –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —É–º–Ω–æ–∂–µ–Ω–∏—è, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ —Ç—Ä–µ—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤. –§—É–Ω–∫—Ü–∏—é –≤—ã–∑–≤–∞—Ç—å —Å –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–º–∏ –Ω–∏–∂–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏. –î–ª—è —Å–ª—É—á–∞—è a4 –≤—ã–±—Ä–∞—Ç—å 3 –ø–µ—Ä–≤—ã—Ö –∏ 3 –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è.
```
a1 = (15, 10 ,5)
a2 = (3, 1)
a3 = [2, 35, 55]
a4 = (5, 10 ,15, 20)
```

–§—É–Ω–∫—Ü–∏—è multiply –ø—Ä–∏–Ω–º–∞–µ—Ç –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ '*args',*  —ç—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞–±–æ—Ç–∞—Ç—å —Å —Ä–∞–∑–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –±–µ–∑ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –æ–ø–∏—Å—ã–≤–∞—Ç—å –≤—Å–µ —ç—Ç–∏ –∞—Ä–≥—É–º–µ–Ω—Ç—ã –≤ —è–≤–Ω–æ–º –≤–∏–¥–µ –ø—Ä–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ —Ñ—É–Ω–∫—Ü–∏–∏.
–í–Ω—É—Ç—Ä–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –∏—Ç–µ—Ä–∞—Ü–∏—è –ø–æ –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º, —á—Ç–æ–±—ã —É–º–Ω–æ–∂–∏—Ç—å –∏—Ö –≤—Å–µ –º–µ–∂–¥—É —Å–æ–±–æ–π –∏ –ø–æ–ª—É—á–∏—Ç—å –∏—Ç–æ–≥–æ–≤—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç. –ï—Å–ª–∏ –∞—Ä–≥—É–º–µ–Ω—Ç —è–≤–ª—è–µ—Ç—Å—è —Å–ø–∏—Å–∫–æ–º –∏–ª–∏ –∫–æ—Ä—Ç–µ–∂–µ–º, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è —É–º–Ω–æ–∂–∞–µ—Ç –ø–µ—Ä–≤—ã–µ –∏ –ø–æ—Å–ª–µ–¥–Ω–∏–µ —Ç—Ä–∏ –∑–Ω–∞—á–µ–Ω–∏—è, –µ—Å–ª–∏ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–ª—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ 6. –ï—Å–ª–∏ –∞—Ä–≥—É–º–µ–Ω—Ç –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å–ø–∏—Å–∫–æ–º –∏–ª–∏ –∫–æ—Ä—Ç–µ–∂–µ–º, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–æ—Å—Ç–æ —É–º–Ω–æ–∂–∞–µ—Ç –µ–≥–æ –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

````python
def multiply(*args):
    result = 1
    for arg in args:
        if isinstance(arg, (list, tuple)):
            values = arg[:3] + arg[-3:] if len(arg) >= 6 else arg
            for value in values:
                result *= value
        else:
            result *= arg
    return result

a1 = (15, 10, 5)
a2 = (3, 1)
a3 = [2, 35, 55]
a4 = (5, 10, 15, 20)

result1 = multiply(a1)
result2 = multiply(*a2)
result3 = multiply(a3)
result4 = multiply(*a4[:-3], *a4[-3:])

print(result1)
print(result2)
print(result3)
print(result4)
````

4. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–¥–∏—É—Å–∞ r –≤—ã—á–∏—Å–ª—è–µ—Ç –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ –∏ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏. –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ—Ä—Ç–µ–∂ –∏–∑ 2 –∑–Ω–∞—á–µ–Ω–∏–π.

–ë—ã–ª–∞ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∞ –±–∏–±–ª–∏–æ—Ç–µ–∫–∞ math –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –≤ —Ñ—É–Ω–∫—Ü–∏–∏ –º–æ–∂–Ω–æ –±—ã–ª–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —á–∏—Å–ª–æ –ø–∏. –í —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –∏ –¥–ª–∏–Ω—ã –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏, –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –±—É–¥—É—Ç –∑–∞–ø–∏—Å–∞–Ω—ã –≤ –∫–æ—Ä—Ç–µ–∂ –∏ —ç—Ç–æ—Ç –∫–æ—Ä—Ç–µ–∂ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è

````python
import math as m

def S_len(r):
    s = m.pi *(r**2)
    l = 2 * m.pi * r
    result = (s,l)
    return result

radius = int(input())
res = S_len(radius)
print(res)
````

5. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —á–µ—Ç—ã—Ä–µ—Ö –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π, –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—â–∏—Ö —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –≤ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ. –°–æ–∑–¥–∞—Ç—å —Å–ª–æ–≤–∞—Ä—å —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –∏–º —Å–∏–º–≤–æ–ª–∞–º–∏ –æ–ø–µ—Ä–∞—Ü–∏–π. –î–ª—è –¥–≤—É—Ö –∑–∞—Ä–∞–Ω–µ–µ –∑–∞–¥–∞–Ω–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª (–Ω–∞–ø—Ä–∏–º–µ—Ä, 25 –∏ 4) –≤—ã–ø–æ–ª–Ω–∏—Ç—å –≤—ã–±—Ä–∞–Ω–Ω—É—é –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫—É—é –æ–ø–µ—Ä–∞—Ü–∏—é.

–û–ø—Ä–µ–¥–µ–ª–µ–Ω—ã —Ñ—É–Ω–∫—Ü–∏–∏ add (—Å–ª–æ–∂–µ–Ω–∏–µ), subtract (–≤—ã—á–∏—Ç–∞–Ω–∏–µ), multiply (—É–º–Ω–æ–∂–µ–Ω–∏–µ) –∏ divide (–¥–µ–ª–µ–Ω–∏–µ) –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —á–µ—Ç—ã—Ä–µ—Ö –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π.
–°–æ–∑–¥–∞–µ—Ç—Å—è —Å–ª–æ–≤–∞—Ä—å operations —Å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ —Å–∏–º–≤–æ–ª–∞–º–∏ –æ–ø–µ—Ä–∞—Ü–∏–π (–∫–ª—é—á–∞–º–∏) –∏ —Ñ—É–Ω–∫—Ü–∏—è–º–∏ (–∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∫–ª—é—á–µ–π), –∫–æ—Ç–æ—Ä—ã–µ –≤—ã–ø–æ–ª–Ω—è—é—Ç —ç—Ç–∏ –æ–ø–µ—Ä–∞—Ü–∏–∏.
–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤–æ–¥–∏—Ç —Å–∏–º–≤–æ–ª –æ–ø–µ—Ä–∞—Ü–∏–∏ –∏ —á–µ—Ä–µ–∑ —Å–ª–æ–≤–∞—Ä—å –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π.

````python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': divide
}

operation = input("–í—ã–±–µ—Ä–∏—Ç–µ –æ–ø–µ—Ä–∞—Ü–∏—é (+, -, *, /): ")

x = 25
y = 4

result = operations[operation](x, y)

print("–†–µ–∑—É–ª—å—Ç–∞—Ç:", result)
````

6. –ù–∞–ø–∏—Å–∞—Ç—å –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä –¥–ª—è —Å—Ç—Ä–æ–∫–æ–≤—ã—Ö –≤—ã—Ä–∞–∂–µ–Ω–∏–π –≤–∏–¥–∞ '<—á–∏—Å–ª–æ> <–æ–ø–µ—Ä–∞—Ü–∏—è> <—á–∏—Å–ª–æ>', –≥–¥–µ <—á–∏—Å–ª–æ> - —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ, –Ω–∞–ø—Ä–∏–º–µ—Ä 113, <–∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è> - –æ–¥–Ω–∞ –∏–∑ –æ–ø–µ—Ä–∞—Ü–∏–π `+,-,*,//,%,^`. –ü—Ä–∏–º–µ—Ä: `calc('13 - 5') -> 8`


–í—ã—Ä–∞–∂–µ–Ω–∏–µ –≤–≤–æ–¥–∏—Ç—Å—è —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é —Ç–∏–ø–∞ —Å—Ç—Ä–æ–∫–∞.
–° –ø–æ–º–æ—â—å—é split —Ä–∞–∑–¥–µ–ª—è–µ–º —Å—Ç—Ä–æ–∫—É –ø–æ –ø—Ä–æ–±–µ–ª–∞–º.
–ó–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ a - –ø–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ, –≤ op - –æ–ø–µ—Ä–∞—Ç–æ—Ä, –≤ b - –≤—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ.
–í —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Ç—Å—è –ø–µ—Ä–µ–±–æ—Ä –≤–æ–∑–º–æ–∂–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤, –ø—Ä–∏ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–∏ —Å op –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ, —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é result –∏ –≤—ã–≤–æ–¥–∏—Ç—Å—è –Ω–∞ —ç–∫—Ä–∞–Ω

````python
def calculator(op, a, b):
    if op == '+':
        result = a + b
    elif op == '-':
        result = a - b
    elif op == '*':
        result = a * b
    elif op == '//':
        result = a // b
    elif op == '%':
        result = a % b
    elif op == '^':
        result = a ** b
    else:
        result = "–ù–µ–≤–µ—Ä–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä"
    print(f"calc('{a} {op} {b}') ->",result)

# –≤–≤–æ–¥ —Å—Ç—Ä–æ–∫–∏
user_input = input("–í–≤–µ–¥–∏—Ç–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 'a op b', –Ω–µ –∑–∞–±—ã–≤–∞—è –ø—Ä–æ –ø—Ä–æ–±–µ–ª—ã): ")

# —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –≤–≤–æ–¥ –Ω–∞ —á–∏—Å–ª–∞ –∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä
parts = user_input.split() #—Ä–∞–∑–¥–µ–ª—è–µ–º –ø–æ –ø—Ä–æ–±–µ–ª–∞–º –≤ —Å–ø–∏—Å–æ–∫
a = float(parts[0])
op = parts[1]
b = float(parts[2])

calculator(op, a, b)
````

–§ –≤—ã–∑—ã–≤–∞–µ–º–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –æ–ø–∏—Å–∞–Ω —Å–ª–æ–≤–∞—Ä—å, –≥–¥–µ –∫–ª—é—á - —ç—Ç–æ —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞, –∞ –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–µ - —Å–∞–º–æ —á–∏—Å–ª–æ.
–ü–µ—Ä–µ–¥–∞–≤–∞–µ–º–∞—è –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ —Å—Ç—Ä–æ–∫–∞, –∫–æ—Ç–æ—Ä—É—é –≤–≤–µ–ª –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å, —Ä–∞–∑–±–∏–≤–∞–µ—Ç—Å—è –Ω–∞ —á–∞—Å—Ç–∏ –ø–æ –ø—Ä–æ–±–µ–ª–∞–º.
–í —Ü–∏–∫–ª–µ —á–∞—Å—Ç—å —Å—Ç—Ä–æ–∫–∏ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç—Å—è —Å –∫–ª—é—á–∞–º–∏ –≤ —Å–ª–æ–≤–∞—Ä–µ –∏ –µ—Å–ª–∏ –µ—Å—Ç—å —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ, —Ç–æ –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ—Ç –∑–∞–ø–∏—Å–∞–Ω–æ –∑–Ω—á–µ–Ω–∏–µ –∫–ª—é—á–∞.

````python
def to_int(text):
    numbers = {
        '–Ω–æ–ª—å': 0, '–æ–¥–∏–Ω': 1, '–¥–≤–∞': 2, '—Ç—Ä–∏': 3, '—á–µ—Ç—ã—Ä–µ': 4, '–ø—è—Ç—å': 5,
        '—à–µ—Å—Ç—å': 6, '—Å–µ–º—å': 7, '–≤–æ—Å–µ–º—å': 8, '–¥–µ–≤—è—Ç—å': 9, '–¥–µ—Å—è—Ç—å': 10,
        '–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å': 11, '–¥–≤–µ–Ω–∞–¥—Ü–∞—Ç—å': 12, '—Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å': 13, '—á–µ—Ç—ã—Ä–Ω–∞–¥—Ü–∞—Ç—å': 14,
        '–ø—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 15, '—à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç—å': 16, '—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 17, '–≤–æ—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 18,
        '–¥–µ–≤—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 19, '–¥–≤–∞–¥—Ü–∞—Ç—å': 20, '—Ç—Ä–∏–¥—Ü–∞—Ç—å': 30, '—Å–æ—Ä–æ–∫': 40,
        '–ø—è—Ç—å–¥–µ—Å—è—Ç': 50, '—à–µ—Å—Ç—å–¥–µ—Å—è—Ç': 60, '—Å–µ–º—å–¥–µ—Å—è—Ç': 70, '–≤–æ—Å–µ–º—å–¥–µ—Å—è—Ç': 80,
        '–¥–µ–≤—è–Ω–æ—Å—Ç–æ': 90, '—Å—Ç–æ': 100, '–¥–≤–µ—Å—Ç–∏': 200, '—Ç—Ä–∏—Å—Ç–∞': 300, '—á–µ—Ç—ã—Ä–µ—Å—Ç–∞': 400,
        '–ø—è—Ç—å—Å–æ—Ç': 500, '—à–µ—Å—Ç—å—Å–æ—Ç': 600, '—Å–µ–º—å—Å–æ—Ç': 700, '–≤–æ—Å–µ–º—å—Å–æ—Ç': 800,
        '–¥–µ–≤—è—Ç—å—Å–æ—Ç': 900
    }

    words = text.lower().split()  
    result = 0

    for word in words:
        if word in numbers:
            result += numbers[word]

    return result

number_text = input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ –≤ –≤–∏–¥–µ —Ç–µ–∫—Å—Ç–∞: ')
number = to_int(number_text)
print(number)
````

8. –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ (–æ—Ç 0 –¥–æ 999) –∏–∑ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ –≤ —á–∏—Å–ª–æ —Ç–∏–ø–∞ int. –ò —Å–æ–æ–±—â–∞–µ—Ç –æ–± –æ—à–∏–±–∫–∞—Ö (–≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –æ–ø–∏—Å–∞–Ω–∏–µ —Ç–∏–ø–∞ –æ—à–∏–±–∫–∏ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∏—Å–ª–æ -1). –ü—Ä–∏–º–µ—Ä: to_int('—Ç—Ä–∏–¥—Ü–∞—Ç—å —Ç—Ä–∏') -> 33 –ü—Ä–∏–º–µ—Ä: to_int("—Å—Ç–æ —Å–æ—Ä–æ–∫ —Ç—Ä–∏–¥—Ü–∞—Ç—å –¥–≤–∞ —Å—Ç–æ") -> -1 –í—ã–≤–æ–¥: —Ç—Ä–∏–¥—Ü–∞—Ç—å - –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ –≤ —á–∏—Å–ª–µ.

````python
def to_int(text):
    num = {
        '–Ω–æ–ª—å': 0, '–æ–¥–∏–Ω': 1, '–¥–≤–∞': 2, '—Ç—Ä–∏': 3, '—á–µ—Ç—ã—Ä–µ': 4,
        '–ø—è—Ç—å': 5, '—à–µ—Å—Ç—å': 6, '—Å–µ–º—å': 7, '–≤–æ—Å–µ–º—å': 8, '–¥–µ–≤—è—Ç—å': 9,
        '–¥–µ—Å—è—Ç—å': 10, '–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å': 11, '–¥–≤–µ–Ω–∞–¥—Ü–∞—Ç—å': 12, '—Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å': 13, '—á–µ—Ç—ã—Ä–Ω–∞–¥—Ü–∞—Ç—å': 14,
        '–ø—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 15, '—à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç—å': 16, '—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 17, '–≤–æ—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 18, '–¥–µ–≤—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 19,
        '–¥–≤–∞–¥—Ü–∞—Ç—å': 20, '—Ç—Ä–∏–¥—Ü–∞—Ç—å': 30, '—Å–æ—Ä–æ–∫': 40, '–ø—è—Ç—å–¥–µ—Å—è—Ç': 50,
        '—à–µ—Å—Ç—å–¥–µ—Å—è—Ç': 60, '—Å–µ–º—å–¥–µ—Å—è—Ç': 70, '–≤–æ—Å–µ–º—å–¥–µ—Å—è—Ç': 80, '–¥–µ–≤—è–Ω–æ—Å—Ç–æ': 90,
        '—Å—Ç–æ': 100, '–¥–≤–µ—Å—Ç–∏': 200, '—Ç—Ä–∏—Å—Ç–∞': 300, '—á–µ—Ç—ã—Ä–µ—Å—Ç–∞': 400, '–ø—è—Ç—å—Å–æ—Ç': 500,
        '—à–µ—Å—Ç—å—Å–æ—Ç': 600, '—Å–µ–º—å—Å–æ—Ç': 700, '–≤–æ—Å–µ–º—å—Å–æ—Ç': 800, '–¥–µ–≤—è—Ç—å—Å–æ—Ç': 900
    }

    words = text.split()

    number = 0
    total = 0
    pred_number = None

    for word in words:
        pre_number = num[word]
        if pred_number is not None and pred_number > pre_number:
            return f"{pre_number} - –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞—è –∑–∞–ø–∏—Å—å —á–∏—Å–ª–∞"

        if pre_number < 10:  # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å–ª–∏ —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ –¥–µ—Å—è—Ç–∏
            if number == 0:  # –ï—Å–ª–∏ –Ω–µ—Ç –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —á–∏—Å–ª–∞, –¥–æ–±–∞–≤–ª—è–µ–º —Ç–µ–∫—É—â–µ–µ
                number += pre_number
            else:  # –ï—Å–ª–∏ –µ—Å—Ç—å –ø—Ä–µ–¥—ã–¥—É—â–µ–µ —á–∏—Å–ª–æ, –¥–æ–±–∞–≤–ª—è–µ–º –∫ –Ω–µ–º—É —Ç–µ–∫—É—â–µ–µ
                number += pre_number
                number *= 10  # –£–º–Ω–æ–∂–∞–µ–º –Ω–∞ 10, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ

        else:
            if number != 0:  # –ï—Å–ª–∏ –µ—Å—Ç—å —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ –¥–µ—Å—è—Ç–∏, –¥–æ–±–∞–≤–ª—è–µ–º –µ–≥–æ –∫ –æ–±—â–µ–º—É —á–∏—Å–ª—É
                total += number
                number = 0  # –°–±—Ä–∞—Å—ã–≤–∞–µ–º —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ –¥–µ—Å—è—Ç–∏

            number += pre_number
            pred_number = pre_number

    total += number
    return total


s1 = input()
result = to_int(s1)
print(result)
s2 = input()
result = to_int(s2)
print(result)
````
